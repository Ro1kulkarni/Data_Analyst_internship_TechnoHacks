```python
#Task 1 : Perform Data Cleaning
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
files="/content/drive/MyDrive/Internship/netflix1.csv"
df=pd.read_csv(files)
numeric_columns=df.select_dtypes(include=[np.number]).columns
z_scores=np.abs((df[numeric_columns]-df[numeric_columns].mean())/df[numeric_columns].std())
df=df[(z_scores<3).all(axis=1)]
plt.figure(figsize=(4,4))
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.title("Box plot of Numeric Columns")
plt.show()
```

```python
#Task 5 : Remove Duplication
import pandas as pd
data={'Name':["Rohan","Rohit","Shreyas","Sahil","Rohan"],
      "Marks":[56,70,66,56,56],
      "City":["Pune","Sangli","Kolhapur","Satara","Pune"]}
df=pd.DataFrame(data)
df
duplicate_rows = df[df.duplicated(keep='first')].value_counts()
df.drop_duplicates(subset="Name",keep="first",inplace=True)
duplicate_rows
df
```

```python
#Task 2 : Calculate summary statistics
import pandas as pd
data={"Name":["Rohan","Rohit","Shreyas","Sahil"],
      "City":["Pune","Sangli","Karad","Satara"],
      "Marks":[55,57,66,61]}
df=pd.DataFrame(data)
mean_value=df["Marks"].mean()
mode_value=df["Marks"].mode().iloc[0]
median_value=df["Marks"].median()
Standerd_daviation_value=df["Marks"].std()
print("Mean Value =",+mean_value)
print("Mode Value =",+mode_value)
print("Median Value =",+median_value)
print("Standerd Deviation Value =",+Standerd_daviation_value)
```

```python
#Task 3 : Visualization using Bar Chart
import pandas as pd
import matplotlib.pyplot as plt
data={"Name":["Rohan","Rohit","Shreyas","Sahil"],
      "City":["Pune","Sangli","Karad","Satara"],
      "Marks":[55,57,66,61]}
df=pd.DataFrame(data)
plt.bar(df['Name'],df['Marks'],color="orange",edgecolor="black")
plt.xlabel('Name')
plt.ylabel('Marks')
plt.title("Bar Chart of Marks")
plt.show()
```

```python
#Task 4 : Pivote Table
import pandas as pd
data={"Name":["Rohan","Rohit","Shreyas","Sahil","Pranav","Pratap"],
      "City":["Sangli","Pune","Satara","Solapur","Kolhapur","Karad"],
      "Marks":[88,66,77,56,72,57]}
df=pd.DataFrame(data)
pivot_table=df.pivot_table(values="Marks",index="Name",columns="City",aggfunc="sum",fill_value=0)
print(pivot_table)
```
